```python
keywords = ['화성시 영천동','화성시 청계동']
filtered_df = df_0[df_0['ADDRESS_JIBUN'].str.contains('|'.join(keywords), na=False)]

filtered_df
```

```python
df=filtered_df[['INCIDENT_ID','INCIDENT_TRAFFIC_CD','LINK_ID','INCIDENT_GRADE_CD', 'START_DATE', 'END_DATE']]
```

```python
df['START_DATE']=df['START_DATE'].astype(str)
df['END_DATE']=df['END_DATE'].astype(str)
```

```python
from datetime import datetime
```

```python
def convert_to_datetime(date_str):
    try:
        datetime_format = '%Y\\년 \\%m\\월 \\%d\\일 \\ %H\\시 \\%M\\분\\'
        date = datetime.strptime(date_str, datetime_format)
        return date
    except ValueError:
        return '2022-01-01 00:00:00'
```

```python
# 데이터 변환

df['START_DATE']=df['START_DATE'].apply(convert_to_datetime)
df['END_DATE']=df['END_DATE'].apply(convert_to_datetime)

df['START_DATE']=pd.to_datetime(df['START_DATE'])
df['END_DATE']=pd.to_datetime(df['END_DATE'])
df['DuringTime']=df['END_DATE']-df['START_DATE']
```

```python
incident_df=df
```

```python
# LINK_ID 결측치 삭제(확인결과 차량 통행에 영향 주지 않는 사고들)

incident_df=incident_df.dropna(subset=['LINK_ID'])
```

```python
#소통등급+돌발등급으로 돌발 영향도 정함
#원활+B등급 -> 원활+A등급 -> 서행+B등급 -> 서행+A등급 -> 지체+B등급 -> 지체+A등급  순서대로 심각도 정함

value_map_GCD={'A0401':1,'A0402':0.5}
value_map_TCD={'A0501':0,'A0502':1,'A0503':2}
value_map_near={1:'N'}
value_map_mid={1:'M'}
value_map_far={1:'F'}
```

```python
merged_df['INCIDENT_TRAFFIC_CD']=merged_df['INCIDENT_TRAFFIC_CD'].map(value_map_TCD)
merged_df['INCIDENT_GRADE_CD']=merged_df['INCIDENT_GRADE_CD'].map(value_map_GCD)
merged_df['near_incident']=merged_df['near_incident'].map(value_map_near)
merged_df['mid_incident']=merged_df['mid_incident'].map(value_map_mid)
merged_df['far_incident']=merged_df['far_incident'].map(value_map_far)
```

```python
merged_df['INCIDENT_LV']=merged_df['INCIDENT_TRAFFIC_CD']+merged_df['INCIDENT_TRAFFIC_CD']
```

```python
merged_df.drop(['INCIDENT_ID','LINK_ID','INCIDENT_TRAFFIC_CD','INCIDENT_GRADE_CD'],axis=1,inplace=True)
merged_df
```

```python
# 이어붙이는 함수 정의
def concat_incidents(row):
    concatenated = ""
    for column in ['INCIDENT_LV', 'near_incident', 'mid_incident', 'far_incident']:
        if not pd.isna(row[column]):
            concatenated += str(row[column])
    return concatenated

# 새로운 컬럼 추가
merged_df['incident_state'] = merged_df.apply(concat_incidents, axis=1)
```

```python
incident_avn=merged_df.drop(columns=['INCIDENT_LV', 'near_incident', 'mid_incident', 'far_incident'])
```

```python
df1 = pd.read_excel('C:\\Users\\ajou\\Desktop\\Data\\dongtan\\for_incidentcode.xlsx')
df2 = incident_avn
```

```python
df1['time']=pd.to_datetime(df1['time'])
df1=df1.set_index('time')
```

```python
# df1과 df2의 'avn' 컬럼 값들을 가져와 avn_list에 저장
avn_list = df2['new avn'].unique()

# 각 avn별로 df2의 정보를 활용하여 df1을 업데이트
for avn in avn_list:
    avn_rows = df2[df2['new avn'] == avn]  # 현재 avn에 해당하는 df2의 행들
    for _, row in avn_rows.iterrows():
        start_time = pd.Timestamp(row['START_DATE'])
        end_time = pd.Timestamp(row['END_DATE'])
        incident_state = row['incident_state']
        
        # start_time과 end_time을 가장 가까운 5분 단위로 반올림
        rounded_start_time = start_time.round('5T')
        rounded_end_time = end_time.round('5T')
        
        # Find the indices corresponding to the rounded times
        start_idx = df1.index[df1.index == rounded_start_time][0]
        end_idx = df1.index[df1.index == rounded_end_time][0]
        
        # Update the incident_state values in df1
        df1.loc[start_idx:end_idx, avn] = incident_state
```

```python
df2=df1.reset_index()
df2=df2.fillna(0)
```
