```python
columns=[40221,  40222,  40223,  40224,  40231,  40232,  40233,  40234,
        40241,  40242,  40243,  40244,  40311,  40312,  40313,  40314,  40321,
        40322,  40323,  40324,  40331,  40332,  40333,  40334,  40341,  40342,
        40343,  40344,  40351,  40352,  40353,  40354,  40371,  40372,  40373,
        40374,  40381,  40382,  40383,  40384,  40391,  40392,  40393,  40394]
```

```python
df_not0=df.copy()
df_not0.iloc[:,1:]=""
df_not0
```

```python
for i in columns:
    for j in range(df.shape[0]):
        if df[i].iloc[j]==0:
            df_not0[i].iloc[j]=link_char[i].iloc[9]
        else: 
            df_not0[i].iloc[j]=df[i].iloc[j]
            
            
df_not0
```

```python
data = {'main': [40221, 40222, 40223, 40224, 40231, 40232, 40233, 40234,
                 40241, 40242, 40243, 40244, 40311, 40312, 40313, 40314, 40321,
                 40322, 40323, 40324, 40331, 40332, 40333, 40334, 40341, 40342,
                 40343, 40344, 40351, 40352, 40353, 40354, 40371, 40372, 40373,
                 40374, 40381, 40382, 40383, 40384, 40391, 40392, 40393, 40394],
        'upstream': [''] * 44
}
df_coordinate = pd.DataFrame(data)
```

```python
df_coordinate['upstream']=[40232,'','',40242,40241,40392,40382,40223,'',40312,40233,40222,'','',40323,40244,40311,'',40333,'',
                          40321,'',40343,40393,40331,'',40353,40394,40341,'','',40374,40391,40352,'',40384,'',40372,'',40231,
                          40332,40342,40373,40234]
```

```python
df_coordinate
```

```python
df_ratio=df.copy()
df_ratio.iloc[:,1:]=''
df_ratio
```

```python
for i in range(df.shape[0]):
    for j in range(df_coordinate.shape[0]):
        try:
            link_m = df_coordinate['main'].iloc[j]
            link_u = df_coordinate['upstream'].iloc[j]
            
            
            df_ratio[link_m].iloc[i] = df_not0[link_m].iloc[i] / df_not0[link_u].iloc[i]
            
        except KeyError:
            df_ratio[link_m].iloc[i] = 0

```

```python
df_ratio

#상류부 링크가 존재하지 않는 경우 : 0
#상류부 링크 속도 값이 0인 경우 : 
#해당 시간에 교통량이 존재하지 않고 nan 값이 없는 것으로 보아 차량이 검지되지 않은 경우라고 가정하여 최고 속도 값으로 대체함
```
